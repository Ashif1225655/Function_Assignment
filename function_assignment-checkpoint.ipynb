```python
#(1).What is the difference between a function and a method in Python?

# Function >> 
#A function is a block of reusable code that is independent and can 
# be called from anywhere in the program. we do not need to attach it to anything specific. 
# For example, print() is a function. You just call it like print("Hello"), and it does its job.

def greet(name):
    return f"Hello, {name}!"

# Call the function
print(greet("Pw skills"))  # Output: Hello, Alice!

# Method >>
#A method is similar to a function, but it's tied to an object or class. In other words,
# it "belongs" to something. For example, my_list.append(5) is calling the append() method on the 
# list object my_list. Methods are functions that operate on the data inside the object they are 
# attached to.

# Class with a method
class Person:
    def __init__(self, name):
        self.name = name

    # Method belonging to the class
    def say_hello(self):
        return f"Hello, my name is {self.name}!"

# Creating an instance (object) of the Person class
person = Person("Bob")

# Calling the say_hello method
print(person.say_hello())  # Output: Hello, my name is Bob!

```

```python
# (4). What is the purpose of the `return` statement in a Python function?

# The return statement in a Python function serves the following main purposes:

# 1. Send a Value Back to the Caller
# The primary role of the return statement is to send a value back to the part of the 
# program where the function was called. This allows the function to produce an output 
# that can be used or stored for later use.

# Example:
def add(a, b):
    return a + b

result = add(3, 4)  # The return value (7) is stored in 'result'
print(result) 


# 2. End the Function Execution
# The return statement also immediately ends the execution of the function. 
# Once the return statement is hit, the function stops running, even if there are more 
# lines of code after it.


#Example:
def check_even(num):
    if num % 2 == 0:
        return True 
    return False 

print(check_even(4))  
print(check_even(3))  


# 3. Return Multiple Values
# A function in Python can return multiple values as a tuple, which can then be unpacked.

# Example:

def get_coordinates():
    return 10, 20

x, y = get_coordinates() 
print(x, y) 

# 4. Returning Nothing
# If a function does not have a return statement, or if the return
# statement is written as return with no value, the function returns None by default.

# Example:

def do_nothing():
    return

result = do_nothing() 
print(result) 
```

```python
# (5). What are iterators in Python and how do they differ from iterables?

# iterators and iterables are both important concepts related to looping through data,
# but they have different roles.

# 1. Iterable
# An iterable is any object that can be looped over or iterated through (like in a for loop).
# Examples of iterables include lists, tuples, strings, sets, and dictionaries. The key feature of
# an iterable is that it has a method called __iter__() which returns an iterator.

# Examples of Iterables:
List: [1, 2, 3]
String: "Hello"
Tuple: (1, 2, 3)

# Example of using an iterable:

my_list = [1, 2, 3]
for item in my_list: 
    print(item)


# 2. Iterator
# An iterator is an object that represents a stream of data. It is created from an iterable
# using the iter() function. An iterator keeps track of its current position in the iterable 
# and produces the next value when you call the next() function. It has two main methods:

# iter() â€“ returns the iterator object itself.
# next() â€“ returns the next item from the sequence or raises StopIteration when there are no more items.
# Example of an Iterator:

my_list = [1, 2, 3]
my_iterator = iter(my_list)  

print(next(my_iterator))  
print(next(my_iterator)) 
print(next(my_iterator)) 
# Here, my_iterator is an iterator created from my_list. Each call to next() 
# moves to the next item in the sequence.
```

```python
# (6). Explain the concept of generators in Python and how they are defined.
# In Python, generators are a special type of function that allow you to iterate over a
# sequence of values lazilyâ€”meaning, they generate values one at a time and only when requested.
# This makes them memory-efficient and ideal for handling large datasets or streams of data.

# Key Characteristics of Generators:
# Yield values lazily: Generators donâ€™t store all the values in memory at once. Instead,
# they produce values on the fly, one at a time, using the yield keyword.
# Pause and Resume: Unlike normal functions that terminate when they reach a return statement,


# generators can pause their execution and resume where they left off when called again.
# State retention: Generators remember their state between successive calls.
# How to Define a Generator
# A generator is defined like a normal function but uses yield instead of return to return values.

# Example of a Simple Generator:

def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # Yield pauses the function and returns a value
        count += 1
# How to Call a Generator:

counter = count_up_to(5)  

# Iterate through the generator using a loop
for num in counter:
    print(num)


# In this example, count_up_to(5) is a generator that will yield numbers from 1 to 5. 
# Each time the yield statement is hit, the generator pauses, returns the value, 
# and resumes from where it left off when called again.
```

```python
# (7). What are the advantages of using generators over regular functions?
# Advantage of using generator
# Using generators in Python offers several advantages over regular functions,
# especially in scenarios involving large datasets or streams of data. 
# Here are the key benefits:

# 1. Memory Efficiency
# Generators produce items one at a time, meaning they do not store the entire sequence
# in memory at once. This makes them very efficient for working with large datasets or 
# infinite sequences.

# Example: A regular function that generates a large list would consume a lot of memory,
# but a generator produces values on demand, saving memory.

def get_large_list(n):
    return [i for i in range(n)]

# Generator version
def generate_large_sequence(n):
    for i in range(n):
        yield i
# If n is very large (say millions), the list created by get_large_list() will consume
# a large amount of memory, whereas generate_large_sequence() will only generate one number
# at a time, consuming much less memory.

# 2. Lazy Evaluation
# Generators implement lazy evaluation, meaning values are produced only when they are needed.
# This is particularly useful when working with streams of data (e.g., reading a file line by line),
# infinite sequences, or real-time data processing.

# Example: Reading a large file line by line:


def read_large_file(file_path):
    with open(file_path) as file:
        for line in file:
            yield line
# This way, the generator reads one line at a time, instead of loading the entire file into memory.

# 3. Improved Performance
# Because generators yield one value at a time, they can start producing output immediately,
# instead of waiting for the entire function to execute. This improves performance,
# especially when the entire result set is not needed at once.

# Example: If you only need the first few elements of a large sequence,
# a generator can start producing them immediately:

def generate_numbers():
    for i in range(1000000):
        yield i

# Get only the first 3 numbers
gen = generate_numbers()
print(next(gen))
print(next(gen)) 
print(next(gen))  
# With a regular function, the entire sequence would need to be computed before returning the result.

# 4. Infinite Sequences
# Generators can handle infinite sequences without running into memory issues,
# since they produce one value at a time.

# Example: An infinite sequence generator:

def infinite_numbers():
    num = 0
    while True:
        yield num
        num += 1
```

```python
# (8). What is a lambda function in Python and when is it typically used?

# A lambda function in Python is a small, anonymous function defined with the lambda keyword.
# Unlike regular functions defined with def, a lambda function is typically used when you need a short,
# simple function for a brief period, especially in situations where defining a full function might be
# overkill.

# Syntax of a Lambda Function:
# A lambda function can have any number of arguments but only one expression. It is written as:

# lambda arguments: expression
# Arguments: Like a normal function, you can pass multiple arguments to a lambda function.
# Expression: A single expression is evaluated and returned. No statements (like loops or conditions)
# are allowed.
# Example:

# Regular function
def add(x, y):
    return x + y

# Lambda function
add_lambda = lambda x, y: x + y

# Calling the lambda function
print(add_lambda(3, 5))  # Output: 8

# Typical Use Cases for Lambda Functions
# Short, Single-use Functions (Anonymous) Lambda functions are useful when you need a simple function
# for a one-time use, often in scenarios where you don't need to assign a name to the function.

# Example with sorted():

points = [(1, 2), (3, 1), (5, -1)]

# Sort by the second item in each tuple
points_sorted = sorted(points, key=lambda x: x[1])
print(points_sorted)  # Output: [(5, -1), (3, 1), (1, 2)]


# Example with map():
numbers = [1, 2, 3, 4]

# Square each number using map and a lambda function
squares = map(lambda x: x ** 2, numbers)
print(list(squares))  # Output: [1, 4, 9, 16]
# Example with filter():

numbers = [1, 2, 3, 4, 5]

# Filter out the even numbers using filter and a lambda function
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4]

# Example with sorted():

words = ['apple', 'banana', 'cherry']

# Sort the list of words by their length
sorted_words = sorted(words, key=lambda word: len(word))
print(sorted_words)  # Output: ['apple', 'cherry', 'banana']
```

```python
# 9. Explain the purpose and usage of the `map()` function in Python.

# map()>> 
# Map function in Python is used to apply a given function to each item of an iterable (like a list,
# tuple, etc.) and return a new iterator with the results. It allows you to transform or process the 
# items in an iterable efficiently without the need for explicit loops.

# Syntax of map():

# map(function, iterable)

# function: 
# A function that is applied to each item in the iterable. This can be a regular function or a lambda function.

# iterable:
# An iterable (like a list, tuple, or string) whose items the function will be applied to.

# The map(): 
# function returns a map object (an iterator), which you can convert into a list, tuple, or other data types if needed.

# Example of using map():

# A function to square a number
def square(x):
    return x * x

# Apply the function to a list of numbers
numbers = [1, 2, 3, 4]
squares = map(square, numbers)

# Convert the result to a list and print it
print(list(squares))  # Output: [1, 4, 9, 16]

# Usage with Lambda Functions:
# Lambda functions are often used with map() because they allow you to define simple,
# one liner functions without needing to create a separate function.

# Example with Lambda:

# Use a lambda function to cube each number in the list
numbers = [1, 2, 3, 4]
cubes = map(lambda x: x ** 3, numbers)

# Convert the result to a list and print it
print(list(cubes))  # Output: [1, 8, 27, 64]

# Multiple Iterables in map():
# You can pass more than one iterable to map(), and it will apply the function to corresponding
# items from both iterables. The function must accept as many arguments as there are iterables.

# Example with Two Iterables:

# A function to add two numbers
def add(x, y):
    return x + y

# Two lists to be added
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

# Apply the add function to corresponding elements of the two lists
result = map(add, numbers1, numbers2)

print(list(result))  # Output: [5, 7, 9]
# In this example, map() applies the add() function to corresponding elements from the two 
# lists (numbers1 and numbers2).
```

```python
# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python
# In Python, the map(), reduce(), and filter() functions are used to process and manipulate sequences (like lists or tuples). Each of them serves a different purpose in functional programming. Here's the key difference between them:

# 1. map() Function:
# Purpose: map() applies a function to each item of an iterable (e.g., list, tuple) and returns a map object (which can be converted into a list, tuple, etc.).
# Usage: Used when you want to transform or modify each element of a sequence based on a function.
# Syntax:

# map(function, iterable)
# Example:

numbers = [1, 2, 3, 4]
squares = map(lambda x: x ** 2, numbers)
print(list(squares))  

# 2. filter() Function:
# Purpose: filter() applies a function to each item of an iterable, and only items for which the function returns True are included in the result.
# Usage: Used when you want to filter or select elements of an iterable based on a condition.
# Syntax:

# filter(function, iterable)
# Example:

numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4, 6]

# 3. reduce() Function:
# Purpose: reduce() applies a function cumulatively to the items of an iterable, reducing the iterable to a single value.
# Usage: Used when you want to perform a cumulative operation (e.g., sum, product) on a sequence. It is part of the functools module, so you need to import it first.
# Syntax:

# from functools import reduce
# reduce(function, iterable)

# Example:

from functools import reduce
numbers = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, numbers)
print(total) 
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

#Solution>>

def is_prime(n):
    # Prime numbers are greater than 1
    if n <= 1:
        return False
    # Check for factors from 2 to sqrt(n)
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Function to check prime numbers between 1 and 200
def check_prime_in_range():
    primes = []
    for num in range(1, 201):
        if is_prime(num):
            primes.append(num)
    return primes

# Example usage
prime_numbers = check_prime_in_range()
print("Prime numbers from 1 to 200:", prime_numbers)
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.
#Solution >>
class FibonacciIterator:
    def __init__(self, num_terms):
        # Initialize the number of terms and starting values for Fibonacci sequence
        self.num_terms = num_terms
        self.current_term = 0
        self.a, self.b = 0, 1  # Starting Fibonacci numbers

    def __iter__(self):
        # The __iter__ method returns the iterator object itself
        return self

    def __next__(self):
        # Stop iteration if the current term exceeds the specified number of terms
        if self.current_term >= self.num_terms:
            raise StopIteration
        # Generate the next Fibonacci number
        fib_number = self.a
        self.a, self.b = self.b, self.a + self.b  # Update values for the next iteration
        self.current_term += 1
        return fib_number

# Example usage
fib = FibonacciIterator(10)  # Generate first 10 terms of the Fibonacci sequence
for number in fib:
    print(number)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
#Solution >>
def powers_of_2(max_exponent):
    """
    Generator function that yields powers of 2 up to 2^max_exponent.
    
    Parameters:
    max_exponent (int): The maximum exponent for which to yield 2^exponent.
    
    Yields:
    int: Powers of 2 up to 2^max_exponent.
    """
    exponent = 0
    while exponent <= max_exponent:
        yield 2 ** exponent
        exponent += 1

# Example usage:
for power in powers_of_2(5):
    print(power)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
#Solution >>
def celsius_to_fahrenheit(celsius):
    """
    Convert Celsius to Fahrenheit.
    
    Parameters:
    celsius (float): Temperature in Celsius.
    
    Returns:
    float: Temperature in Fahrenheit.
    """
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 40]

# Use map to apply celsius_to_fahrenheit to each element in the list
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string
#Solution >>
def remove_vowels(s):
    """
    Remove all vowels from a given string.
    
    Parameters:
    s (str): The input string from which vowels are to be removed.
    
    Returns:
    str: The string with vowels removed.
    """
    vowels = 'aeiouAEIOU'
    # Use filter to remove vowels
    filtered_chars = filter(lambda char: char not in vowels, s)
    # Join the filtered characters into a new string
    return ''.join(filtered_chars)

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)

print(result)
```
